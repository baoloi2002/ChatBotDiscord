```python
pattern = r'([a-zA-Z\s]+):(.+)'
```

```python
data = {
    'name': [],
    'line': []
}
```

```python
import csv
```

```python
# Read the CSV file
with open('hp_script.csv', 'r') as file:
    reader = csv.DictReader(file)
    for row in reader:
        data['name'].append(row['character_name'])
        data['line'].append(row['dialogue'])
```

```python
df = pd.DataFrame(data)
```
